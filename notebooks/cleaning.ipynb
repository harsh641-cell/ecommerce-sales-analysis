```python
import pandas as pd

# reading csv file
df = pd.read_csv("messy_ecommerce_data.csv")
```

```python
# checking info about dataframe
df.info()
```

```python
# checking and cleaning colums

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
# checking null values

df.isnull().sum()
```

```python
# checking shape 

df.shape
```

```python
# dropping duplicate

df = df.drop_duplicates()
df.shape # 5 lines of code is been dropped  
```

```python
# filling missing values in review rating

df["review_rating"] = df["review_rating"].fillna(2.5) # a avg rating

# checking did it worked or not

df.isnull().sum()
```

```python
# cleaning string values

df["customer_name"] = df["customer_name"].str.lower().str.strip().str.replace(" " , "_")
df["payment_mode"] = df["payment_mode"].str.lower().str.strip().str.replace(" " , "_")
df["is_shipped"] = df["is_shipped"].str.lower().str.strip().str.replace(" " , "_")
df["category_name"] = df["category_name"].str.lower().str.strip().str.replace(" " , "_")

df.head(20)
```

```python
# replacing 99.0 value with 5 in review rating

df["review_rating"] = df["review_rating"].replace(99.0,5)

# filling nan value in payment mode with cash

df["payment_mode"] = df["payment_mode"].fillna("cash")

# replacing nan values into " "

df["qty"] = df["qty"].fillna(0)

# replacing category name into accessories for nan values

df["category_name"] = df["category_name"].fillna("accessories")

```

```python
# droping negative values in price

# 1. Find the rows where price is negative
negative_prices_index = df[df["unit_price_($)"] < 0].index

# 2. Drop them using their index labels
df = df.drop(negative_prices_index)


df.sample(20)
```

```python
# cleaned unit price 

missing_values_df = df[df["unit_price_($)"].isna()]

df = df.drop(missing_values_df.index)
 
missing_values_df.to_csv("null_prices_index.csv" )
```

```python
# cleaning is shipped

df["is_shipped"] = df["is_shipped"].replace("n" , "no")

df["is_shipped"] = df["is_shipped"].fillna("no")
```

```python
# finnally dropping data of unknown person so we can focus on this clean data

df = df.dropna()

df.to_csv("cleaned_ecommerce_data.csv" , index=False)

# rate this cleaning process out of 10
```

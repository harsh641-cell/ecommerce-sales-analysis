```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading dataset

df = pd.read_csv(r"D:\harsh attrishi\data analysis projects\ecommerce-sales-analysis\processed data\cleaned_ecommerce_data.csv")
df.sample(10)
```

```python
# most used payment mode

data_of_payment_mode = df["payment_mode"].value_counts()

plt.figure(figsize=(10, 6))
sns.set_context("notebook")
sns.barplot(data = data_of_payment_mode , palette = "viridis" )
plt.ylabel("no of their users")
plt.xlabel("payment mode")
plt.tight_layout()

print("most used payment mode")
plt.show()

```

```python
# regular customer

regular_customer = df["customer_name"].value_counts()

plt.figure(figsize=(10, 6))
sns.set_context("notebook")
sns.barplot(data =  regular_customer , palette = "Set2"  )
sns.set_style("darkgrid")
plt.ylabel("customer address")
plt.xlabel("customer name")
plt.tight_layout()

print("most trusted customers")
plt.show()
```

```python
# most selled product

product_qty = df.groupby("product__name")["qty"].sum()

plt.figure(figsize=(10,6))
sns.barplot(data = product_qty , palette="Dark2")
plt.ylabel("ordered quantity")
print("most celled product")
plt.show()
```

```python
# total sales

df["total_price"] = df["qty"] * df["unit_price_($)"]
total_sales = df["total_price"].sum()

print("total sales")
print(total_sales)
df.to_csv("cleaned_ecommerce_data.csv")
```

```python
# sales by category 

category_sales = df.groupby("category_name")["total_price"].sum()

plt.figure(figsize=(10,6))
sns.barplot(data=category_sales , palette="Set1")
plt.grid(color = "grey")
sns.set_style("darkgrid")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.ticklabel_format(style='plain', axis='y')

print("sales by category")
plt.show()
```

```python
# avg rating in both cases shipped or un shipped

avg_rating = df.groupby("is_shipped")["review_rating"].mean()
print("avg rating while")
print(avg_rating)
```

```python
# our regular customers prefer which payment mode

payment_mode_of_customers = df.groupby("customer_name")["payment_mode"].value_counts()

print("most used payment mode by our customers")
print(payment_mode_of_customers)
```
